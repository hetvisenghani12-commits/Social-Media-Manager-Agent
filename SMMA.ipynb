```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "Post_ID": range(1, 16),
    "Post_Type": ["Image", "Video", "Text"] * 5,
    "Likes": np.random.randint(50, 500, 15),
    "Comments": np.random.randint(5, 100, 15),
    "Hour_Posted": np.random.randint(0, 24, 15)
}

df = pd.DataFrame(data)
df
```

```python
df.to_csv("post_history.csv", index=False)
```

```python
df["Engagement_Score"] = df["Likes"] + (df["Comments"] * 2)
df
```

```python
content_analysis = df.groupby("Post_Type")["Engagement_Score"].mean()
content_analysis
```

```python
content_analysis.plot(kind="bar", title="Best Performing Content Type")
plt.show()
```

```python
time_analysis = df.groupby("Hour_Posted")["Engagement_Score"].mean()
best_hour = time_analysis.idxmax()

print("Best hour to post:", best_hour)
```

```python
future_posts = {
    "Scheduled_Post_Type": ["Image", "Video", "Text"],
    "Scheduled_Hour": [best_hour, best_hour+1, best_hour+2]
}

schedule_df = pd.DataFrame(future_posts)
schedule_df
```

```python
schedule_df.to_csv("scheduled_posts.csv", index=False)
```
